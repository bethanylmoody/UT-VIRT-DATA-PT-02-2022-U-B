```python
# Dependencies
import requests

url = "http://www.omdbapi.com/?apikey=trilogy&t="

movies = ["Aliens", "Sing", "Moana"]
```

```python
movie_data = []
for movie in movies:
    data = requests.get(url + movie).json()
    print(data["Director"])
    movie_data.append(data)
movie_data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(movie_data)
```
